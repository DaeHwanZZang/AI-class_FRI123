## 1. 서포트 벡터 머신 (Support Vector Machine, SVM)

데이터를 분류하는 **최적의 결정 경계(Hyperplane)**를 찾는 강력한 알고리즘입니다.

*   **결정 경계 (Hyperplane):** 데이터를 나누는 기준선(2차원) 또는 초평면(고차원)
*   **마진 (Margin):** 결정 경계와 가장 가까운 데이터 포인트 사이의 거리. SVM은 이 마진을 **최대화**하는 것을 목표로 함
*   **서포트 벡터 (Support Vectors):** 경계를 결정하는 데 결정적인 역할을 하는 가장 가까운 데이터 포인트들
*   **커널 트릭 (Kernel Trick):** 직선으로 나눌 수 없는 데이터를 고차원으로 보내 선형 분리가 가능하게 함
*   **슬랙 변수 (Slack Variables):** 약간의 오차를 허용하여 '소프트 마진'을 구현함으로써 노이즈에 강하게 함

## 2. 딥러닝 학습의 5단계 (The Training Loop)

인공신경망이 데이터를 학습하는 순환 구조입니다.

1.  **순전파 (Forward Propagation):** 데이터를 층(Layer)에 통과시켜 예측값($\hat{y}$) 계산
2.  **손실 함수 계산 (Loss Function):** 정답과 예측값의 오차 계산 (예: MSE, Cross-Entropy)
3.  **역전파 (Backward Propagation):** 오차를 뒤로 전달하며 각 가중치($W$)가 오차에 기여한 정도(기울기) 계산
4.  **최적화 (Optimizer / Weight Update):** 경사하강법 등을 사용하여 가중치를 실제로 수정
5.  **반복 (Epochs / Iteration):** 위 과정을 전체 데이터에 대해 여러 번 반복

### 신경망의 5대 구성 요소
*   **가중치 (Weights):** 데이터의 중요도
*   **편향 (Biases):** 결과값에 더해지는 상수
*   **활성화 함수 (Activation Function):** 비선형성을 부여 (ReLU, Sigmoid 등)
*   **층 (Layers):** 입력층, 은닉층, 출력층의 구조
*   **학습률 (Learning Rate):** 가중치 업데이트 보폭

## 3. 비선형성과 역전파의 메커니즘

*   **비선형성 부과 (Forward):** 층 사이에 활성화 함수를 넣어 복잡하고 굽어지는 패턴을 학습할 수 있게 함
*   **역전파 영향 (Backward):** 활성화 함수의 미분값에 따라 에러 정보(기울기)가 잘 전달되느냐, 중간에 사라지느냐(기울기 소실 문제)가 결정됨

## 4. 경사하강법 (Gradient Descent) 실습

**공식:** 
$$
W_{new} = W_{old} - \eta \times \frac{\partial Loss}{\partial W}
$$

### 🔍 예제 계산 ($y = (x-3)^2 + 10$)
*   **기울기:** $2x - 6$, **학습률($\eta$):** $0.2$
*   **Step 1:** $x=10$ 일 때, 기울기 $14 \rightarrow$ 다음 $x = 10 - (0.2 \times 14) = \mathbf{7.2}$
*   **Step 2:** $x=7.2$ 일 때, 기울기 $8.4 \rightarrow$ 다음 $x = 7.2 - (0.2 \times 8.4) = \mathbf{5.52}$

## 5. 지니 계수 (Gini Index) - 의사결정나무의 기준

노드 내 데이터의 **불순도(Impurity)**를 측정하는 지표입니다.

### 수식
$$
G = 1 - \sum_{i=1}^{n} (p_i)^2
$$
($p_i$: 특정 클래스가 나타날 확률)

*   **0 (최소):** 완벽히 순수함 (한 종류만 있음)
*   **0.5 (최대):** 가장 불순함 (반반 섞임)
*   **목표:** 모델은 이 지니 계수가 **낮아지는 방향**으로 데이터를 분할함

## 6. 앙상블 기법 (Ensemble Methods)

"여러 개의 약한 모델을 합쳐 하나의 강력한 모델을 만드는 집단지성 기법"

1.  **배깅 (Bagging):** 데이터를 무작위 샘플링(Bootstrap)하여 여러 모델을 **병렬**로 학습 (예: 랜덤 포레스트)
2.  **부스팅 (Boosting):** 앞선 모델의 **실수(오차)를 보완**하며 **순차적**으로 학습 (예: XGBoost, LightGBM)
3.  **스태킹 (Stacking):** 여러 모델의 예측 결과를 다시 입력 데이터로 삼아 **최종 모델**이 재학습

## 7. 부트스트랩 (Bootstrap)

*   **정의:** 원본 데이터에서 **복원 추출**을 통해 새로운 샘플 데이터셋을 만드는 기법
*   **OOB (Out-of-Bag):** 샘플링에서 제외된 약 36.8%의 데이터. 테스트 데이터 대용으로 활용 가능
*   **장점:** 데이터가 한정적일 때 모델의 다양성을 확보하고 과적합(Overfitting)을 방지함

```python
# 1. 경사하강법 시뮬레이션 코드
def gradient_descent_demo():
    current_x = 10
    lr = 0.2
    print("--- Gradient Descent Update ---")
    for i in range(1, 4):
        grad = 2 * current_x - 6
        next_x = current_x - (lr * grad)
        print(f"Step {i}: x={current_x:.2f} -> grad={grad:.2f} -> next_x={next_x:.2f}")
        current_x = next_x

# 2. 지니 계수 계산 코드
def gini_index(counts):
    total = sum(counts)
    p_sum_sq = sum([(c/total)**2 for c in counts])
    return 1 - p_sum_sq

gradient_descent_demo()
print(f"\n지니 계수 (빨강 5, 파랑 5): {gini_index([5, 5])}")
print(f"지니 계수 (빨강 10, 파랑 0): {gini_index([10, 0])}")
```
